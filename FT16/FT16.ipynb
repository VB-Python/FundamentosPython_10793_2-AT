```python
#Comecemos por criar um gráfico simples. Vamos traçar duas listas contendo coordenadas X, Y para o gráfico.

import matplotlib.pyplot as plt
```

```python
 # initializing the data
 x = [10, 20, 30, 40]
 y = [20, 30, 40, 50]
 # plotting the data
 plt.plot(x, y)
 # Adding the title
 plt.title("Simple Plot")
 # Adding the labels
 plt.ylabel("y-axis")
 plt.xlabel("x-axis")
 plt.show()
```

```python
x = [1,2,3]
y = [2,4,1]

plt.plot(x, y)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('My first graph!')

plt.show()
```

```python
# Python program to show pyplot module

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis((0, 6, 0, 20))
plt.show()
```

```python

year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]
#with different colored labels of two countries

plt.plot(year, e_india, color ='orange', marker = "o", label ='India')

plt.plot(year, e_bangladesh, color ='g', label ='Bangladesh')
plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')
plt.title('Electricity consumption per capital of India and Bangladesh')

plt.legend() 
plt.show() 
```

```python

year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]
#with different colored labels of two countries

plt.plot(year, e_india, color ='orange',marker = "o", label ='India')

plt.plot(year, e_bangladesh, color ='g', linestyle='--', label ='Bangladesh')
plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')
plt.title('Electricity consumption per capital of India and Bangladesh')

plt.legend()
plt.show() 
```

```python
#11. Crie um dicionário Python de 3 produtos. As chaves do dicionário podem ser nomes de
#alimentos e os valores podem ser preços

produtos = {"Pão": 10,
            "Leite": 8,
            "Sorverte": 12}

```

```python
#12. Crie um gráfio de barras onde x-axis recebe como valor as chaves do dicionário e y-axis recebe
#os valores do dicionário

fig, ax = plt.subplots()
ax.bar(produtos.keys(), produtos.values())
```

```python
#13. Adicione um título, xlabel e ylabel ao último plot criado

fig, ax = plt.subplots()
ax.bar(produtos.keys(), produtos.values())
ax.set(title="Lista de produtos", ylabel="Preço");
```

```python
#14. Faça o mesmo plot, dessa vez com barras horizontais

fig, ax = plt.subplots()
ax.barh(produtos.keys(), produtos.values())
ax.set(title="Lista de produtos", xlabel="Preço");
```

```python
#11. Crie um dicion√°rio Python de 3 produtos. As chaves do dicion√°rio podem ser nomes de
#alimentos e os valores podem ser pre√ßos

produtos = {"P√£o": 10,
            "Leite": 8,
            "Sorverte": 12}

```

```python
#12. Crie um gr√°fio de barras onde x-axis recebe como valor as chaves do dicion√°rio e y-axis recebe
#os valores do dicion√°rio

fig, ax = plt.subplots()
ax.bar(produtos.keys(), produtos.values())
```

```python
#13. Adicione um t√≠tulo, xlabel e ylabel ao √∫ltimo plot criado

fig, ax = plt.subplots()
ax.bar(produtos.keys(), produtos.values())
ax.set(title="Lista de produtos", ylabel="Pre√ßo");
```

```python
#14. Fa√ßa o mesmo plot, dessa vez com barras horizontais

fig, ax = plt.subplots()
ax.barh(produtos.keys(), produtos.values())
ax.set(title="Lista de produtos", xlabel="Pre√ßo");
```
